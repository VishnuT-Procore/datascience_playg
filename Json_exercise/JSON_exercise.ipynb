## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

## JSON exercise

Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,
1. Find the 10 countries with most projects
2. Find the top 10 major project themes (using column 'mjtheme_namecode')
3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
# read world_bank_projects.json to local varaible.
worldbank_projects_df = pd.read_json('data/world_bank_projects.json')
```

```python
#check for any mising value on 'countryshortname'
worldbank_projects_df.countryshortname.describe()
```

```python
# empty values 
worldbank_projects_df.countryshortname[worldbank_projects_df.countryshortname == ''].count()
```

```python
# null values 
worldbank_projects_df[worldbank_projects_df.countryshortname.isnull()].count()
```

```python
# group by based on country short code
worldbank_projects_grouped = worldbank_projects_df.groupby(worldbank_projects_df.countryshortname)['_id']
```

```python
# There are some projects awarded to group of countries as regional initative. e.g. 
# worldbank_projects_df[worldbank_projects_df.countryshortname == 'Africa']. 
# I am ignoring for sake of simplicity.
worldbank_projects_grouped.count().sort_values(ascending=False).head(10)
```

```python
# Find the top 10 major project themes (using column 'mjtheme_namecode')
# load json as list
world_bnk_json_as_list = json.load(open('data/world_bank_projects.json'))

#normalize theme from given JSON
norm_world_bnk = json_normalize(world_bnk_json_as_list,'mjtheme_namecode')

#group by based on name, sort based on count and select top 10 results.
norm_world_bnk.groupby('name')['name'].count().sort_values(ascending=False).head(10)
```

```python
'''
Loop thru' only rows which have missing name.

If Name is missing for a row then look for name for same code.
'''
def fill_missing(row):
    p = str(row['code'])
    if (row.name):
        row['name'] = norm_world_bnk[(norm_world_bnk.code == p) & (norm_world_bnk.name != '')].name.unique()[0]
    return row
```

```python
# for each row in norm_world_bnk, appply fill missing method.
norm_world_bnk[['code','name']].apply(lambda row : fill_missing(row), axis = 1)
```
