```python
#Include mathplot library  
#mathplot is a data visualization tool with heavy relience on Numpy arrays
#Pyplot is simple ploting interface with in mathplot library, with similaritirs to MATH Lab like interface. 
import matplotlib.pyplot as plt 
import matplotlib
```

```python
#Include numpy library 
#Numpy is library for the Python programming language, adding support for large, multi-dimensional arrays and matrices,
#along with a large collection of high-level mathematical functions to operate on these arrays#.
import numpy as np
```

```python
x = np.linspace(0,5,num=10)
#Returns num evenly spaced samples, calculated over the interval [start, stop].
y = x ** 2
```

```python
#matplotlib.pyplot.plot(*args, **kwargs) 
plt.plot(x,y,'r') #r means red color
plt.show()
```

```python
#subplot(nrows, ncols, plot_number)
#Where nrows and ncols are used to notionally split the figure into nrows * ncols sub-axes, 
#and plot_number is used to identify the particular subplot that this function is to create within the notional grid. 
#plot_number starts at 1, increments across rows first and has a maximum of nrows * ncols.
plt.subplot(1,2,1)
```

```python
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(y,x,'g*-')
plt.show()
```

```python
#Object oriented mathplotAPI 
fig = plt.figure() 

#Add an axes at position rect [left, bottom, width, height] where all quantities are in fractions of figure width and height 
#If the figure already has an axes with the same parameters, then it will simply make that axes current and return it
axes = fig.add_axes([0.1,0.1,0.8,0.8]) 

axes.plot(x,y,'r')
axes.set_xlabel('x axes')
axes.set_ylabel('y axes')
axes.set_title('title');
plt.show()
```

```python
fig = plt.figure()

#rect [left, bottom, width, height] 
axes_main = fig.add_axes([0.1,0.1,0.8,0.8])  #main axes 
axes_sub = fig.add_axes([0.2,0.5,0.4,0.3]) # sub axes 

axes_main.plot(x,y,'r')
axes_sub.plot(x,y,'g^-')

plt.show()
```

```python
fig, axes = plt.subplots()

axes.plot(x,y,'r')
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('title')
plt.show()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

for ax  in axes:
    ax.plot(x,y,'r')

# This method, re-arranges anyoverlaping text between different subplots lables.
#refer: https://matplotlib.org/users/tight_layout_guide.html
fig.tight_layout()

plt.show()
```

```python
# This statement informs system to generate 800X400 picture with 100 dots per inch figure
fig,axes = plt.subplots(figsize=(8,4),dpi=100)

axes.plot(x,y,'r')
axes.set_xlabel('x lab')
axes.set_ylabel('y lab')

plt.show()
```

```python
fig, axes = plt.subplots()


axes.plot(x, x**2, label = r'$ y = \alpha^2$')
axes.plot(x,x**3,label = r'$ y = \alpha^3$')
axes.set_xlabel(r'$\alpha$')
axes.set_ylabel(r'$y$')
axes.legend()
plt.show()
```

```python
#Setting colors 
fig, axes = plt.subplots()


axes.plot(x, x**2, 'r^-',label = r'$ y = \alpha^2$',)
axes.plot(x,x**3,'g-',label = r'$ y = \alpha^3$')
axes.set_xlabel(r'$\alpha$')
axes.set_ylabel(r'$y$')
axes.legend()
plt.show()
```

```python
#Custom plot range 
fig, axes = plt.subplots(1,3,figsize=(12,4))

axes[0].plot(x,x**2,x,x**3)
axes[0].set_title("default curve")

axes[1].plot(x,x**2,x,x**3)
axes[1].axis('tight')
axes[1].set_title("tight curve")

axes[2].plot(x,x**2,x,x**3)
axes[2].set_ylim([0,60])
axes[2].set_xlim([2,5])
axes[2].set_title('custom axes range')

plt.show()
```

```python
#placement of custom ticks and labels 

fig,axes = plt.subplots(figsize=(10,4))

axes.plot(x,x**2,x,x**3)

axes.set_xticks([1,2,3,4,5])
axes.set_xticklabels([r'$alpha$',r'$beta$',r'$gamma$',r'$delta$',r'$epsilon$'],fontsize=18)

axes.set_yticks([0,50,100,150])
plt.show()
```

```python
#Axis Number and Cell spacing 
matplotlib.rcParams['xtick.major.pad'] = 5
matplotlib.rcParams['ytick.major.pad'] = 5


fig,ax = plt.subplots(1,2)

ax[0].plot(x,x**2,x,np.exp(x))
ax[0].set_yticks([0,50,100,150])

ax[0].yaxis.lablepad = 5 
ax[0].xaxis.lablepad = 5 

ax[0].set_xlabel('x')
ax[0].set_ylabel('y')


ax[1].plot(x,x**2,x,np.exp(x))
ax[1].set_yticks([0,50,100,150])

ax[1].set_xlabel('x')
ax[1].set_ylabel('y')

fig.tight_layout()
plt.show()
```

```python
fig,ax = plt.subplots(1,2,figsize=(10,3))

# default grid
ax[0].plot(x,x**2,x,x**3,x,np.exp(x),lw=10)
ax[0].grid(True)
ax[0].legend(['square','cude','expo'])

# custom grid
ax[1].plot(x,x**2,x,x**3,x,np.exp(x),lw=10)
ax[1].grid(color='b',alpha=0.5,linewidth=0.5,linestyle='dashed')
ax[1].legend(['square','cude','expo'])

plt.show()
```

```python
fig,ax1 = plt.subplots()

ax1.plot(x,x**2,lw=2,color='blue')
ax1.set_ylabel(r"area $(m^2)$",fontsize=12,color = 'blue')
for label in ax1.get_yticklabels():
    label.set_color('blue')

ax2 = ax1.twinx()
ax2.plot(x,x**3,lw=3,color='red')
ax2.set_ylabel(r"Volume $(m^3)$",fontsize=12,color='red')

for label in ax2.get_yticklabels():
    label.set_color('red')
    
plt.show()
```

```python
fig,ax = plt.subplots()

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))

ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

xx = np.linspace(-0.7,1.0,100)
ax.plot(xx,xx**3)

plt.show()
```

```python
fig,ax = plt.subplots(1,4,figsize=(12,4))


ax[0].scatter(xx,xx + 0.25*np.random.randn(len(xx)))
ax[0].set_title('scatter plot')

ax[1].step(x,x**2,lw=2)
ax[1].set_title('step function')

ax[2].bar(x,x**2,align="center",width=0.5,alpha=0.5)
ax[2].set_title('bar function')

ax[3].fill_between(x,x**2,x**3,color='red',alpha=1)
ax[3].set_title('fill in between')

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0.0,0.0,0.6,0.6],polar=True)
t = np.linspace(0,2*np.pi,100)

ax.plot(t,t,color='blue',lw=3)

plt.show()
```

```python
x = np.random.randn(10000)

fig,ax = plt.subplots(1,2,figsize=(12,4))

ax[0].hist(x)
ax[0].set_title('histogram')
ax[0].set_xlim((min(x),max(x)))

ax[1].hist(x,cumulative=True,bins=50)
ax[1].set_title('cummulative detailed histogram')
ax[1].set_xlim((min(x),max(x)))

plt.show()

```
