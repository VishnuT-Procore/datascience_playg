```python
# Observe total crimes vs. violent crimes #
# A violent crime is defined as a crime where there is an arrest#
# I use area plot as it better captures the relative aspect of subject.
```

```python
crimes_2015_today_orig.head()
```

```python
# Calculate violent crime and violent + non-violent crime
crimes_2015_violent_crime = crimes_2015_today_orig.copy()
crimes_2015_violent_crime['arrest_crime_count'] = ''
crimes_2015_violent_crime['total_crime_count'] = ''
crimes_2015_violent_crime['arrest_crime_count'] = crimes_2015_violent_crime[crimes_2015_violent_crime.Arrest == True].groupby(pd.TimeGrouper('D')).count()
crimes_2015_violent_crime['total_crime_count'] = crimes_2015_violent_crime.groupby(pd.TimeGrouper('D')).count()
```

```python
crimes_2015_violent_crime.loc['2015-11-01':,'total_crime_count']
```

```python
# graph area plot  violent_crime_count vs total_crime_count over time.
fig, ax = subplots()
crimes_2015_violent_crime.plot(x=crimes_2015_violent_crime.index, y=['total_crime_count','arrest_crime_count'],ax=ax,stacked=False,figsize=[16,6],kind='area',fontsize=40,colormap='autumn',legend=True)
ax.legend(fontsize='xx-large')
```

### Observations ###
####Violent crime trends with overall crime####

```python
# People generally divide area into beats to monitor city
# lets see which contributes to more crime to city
```

```python
crime_per_beat_id  = crimes_2015_today_orig.groupby('Beat').count()
```

```python
crime_per_beat_id['mean_value'] = ''
crime_per_beat_id['mean_value'] = crime_per_beat_id['ID'].mean()
crime_per_beat_id.head()
```

```python

fig, ax = plt.subplots()
sns.barplot(y='ID', x='Beat' ,ax=ax,  data = crime_per_beat_id.reset_index())
#mean line 
sns.pointplot(data=crime_per_beat_id, x=crime_per_beat_id.index, y='mean_value',hue="mean_value",linestyles=["--"],ci=None)
#plt.plot(y= 'mean_value', data=crime_per_beat_id,linewidth=2)
plt.show()
```

```python
crime_per_beat_id_morethan_average = crime_per_beat_id.ID > crime_per_beat_id['mean_value']
```

```python
crime_per_beat_id[crime_per_beat_id_morethan_average].count()
crime_per_beat_id.count()
```
